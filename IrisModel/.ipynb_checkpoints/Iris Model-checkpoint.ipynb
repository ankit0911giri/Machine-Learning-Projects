```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier


rfc = RandomForestClassifier(random_state = 1)
```

```python
data = load_iris()
```

```python
# convert data into dataframe

irisData = pd.DataFrame(data.data , columns = data.feature_names)
```

```python
x = irisData.drop("target",axis = 1)
y = irisData["target"]
```

```python
from sklearn.model_selection import train_test_split

train_x,val_x,train_y,val_y = train_test_split(x,y,random_state = 1)

rfc.fit(train_x,train_y)

rfc.score(val_x,val_y)
```

```python
from sklearn.metrics import mean_absolute_error


def getMae(maxleaf, train_x,val_x,train_y,val_y):
    rfc = RandomForestClassifier(max_leaf_nodes = maxleaf,random_state = 1)
    rfc.fit(train_x,train_y)
    predictedVal = rfc.predict(val_x)
    print(rfc.score(val_x,val_y) , mean_absolute_error(predictedVal,val_y) , maxleaf)
```

```python
nodeValues = [50,100,200,300,400,500]

for i in nodeValues:
    getMae(i,train_x,val_x,train_y,val_y)
    
```

```python
movies = pd.read_csv("movies.csv")
movies.head(5)
```

```python
new_list = ["title","director","cast","listed_in","description"]
movies = movies[new_list]
movies.head(5)
```

```python
movies["information"] = movies["director"] + movies["cast"] + movies["listed_in"] + movies["description"]
movies.head(5)
```

```python
new_list = ["director","cast","listed_in","description"]
movies.drop(new_list,axis = 1, inplace = True)
movies.head(5)
```

```python
movies.information = movies.information.str.lower()
```

```python
movies.head(5)
```

```python
movies.information = movies["information"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies["information"] = movies.information.apply(lambda x:"".join(x))
```

```python
movies.head(5)
```

```python
movies.information.iloc[0]
```

```python
import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stemming(text):
    words = []
    for i in text.split():
        words.append(ps.stem(i))
    return " ".join(words)
```

```python
movies.information = movies.information.apply(stemming)
```

```python
movies.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5500, stop_words = "english")
```

```python
vectors = cv.fit_transform(movies["information"]).toarray()
```

```python
vectors
```

```python
vectors[0]   # this is our first movie information in array form
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar_movies = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similar_movies[0])),reverse = True,key = lambda x: x[1])[1:6]
```

```python
movies["title"].iloc[333]
```

```python
def recommended(name):
    movie_index = movies[movies["title"] == name].index[0]
    distance = similar_movies[movie_index]
#     movies_list = sorted(list(enumerate(similar_movies[0])), reverse = True, key = lambda x: x[1])[1:6]
    movies_list = sorted(list(enumerate(distance)),reverse = True,key = lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(movies["title"].iloc[i[0]])
```
